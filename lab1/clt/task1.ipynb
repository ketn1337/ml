```python
%matplotlib inline
import matplotlib.pyplot as plt
import scipy.stats as sts
import numpy as np
```

Будем рассмотривать экспоненциальное распределение случайно величины с  λ=1 . Плотность вероятности для экспоненциального распределения будет иметь следующий вид:

$
f_X(x) = \begin{cases}
\lambda \,e^{-\lambda x} ,& x \ge 0, \\
0 ,& x < 0.
\end{cases}.
$

Сгенерируем из него выборку объёма 1000 и сравним гистограмму выборки и с теоретической плотностью распределения случайной величины.

```python
rv = sts.expon()
sample = rv.rvs(1000)

x = np.linspace(0, 15, 1000)
pdf = rv.pdf(x)
```

Построим гистограмму выборки и наложим на нее теоретическую оценку плотности вероятности экспоненциального распределения.

```python
plt.hist(sample, 20, density=True)
plt.plot(x, pdf, 'red')
```
