```python
import pandas as pd
import numpy as np
from scipy import spatial
import re
```

Считаем данные из файла sentences.txt. Далее приведем их к нижнему регистру и разделим на слова, а также удалим пустые строки.

```python
words_by_sentences = []

with open("sentences.txt") as f:
    words_by_sentences = [re.split('[^a-z]', i.lower()) for i in f]
    for i in range(0, len(words_by_sentences)):
        words_by_sentences[i] = [j for j in words_by_sentences[i] if j]

    f.close()
        
print(words_by_sentences)
```

Создадим словарь, в котором ключ будет словом из множества data, а значение будет варьироваться от 0 до (d - 1), где d - число различных слов в предложениях.

```python
data = set()
for sentence in words_by_sentences:
    for i in sentence:
        data.add(i)

unique_words = list(data)
data_dict = dict()

for i in range(0, len(unique_words)):
    data_dict.update({unique_words[i]: i})

data_dict
```

Создадим матрицу размера n x d, где n - число предложений. Заполним ее: элемент с индексом (i, j) будет равен количеству вхождений j-го слова в i-е предложение.

```python
matrix = np.zeros((22, 254))

for i in range(0, len(words_by_sentences)):
    for j in words_by_sentences[i]:
        matrix[i, data_dict[j]] += 1

matrix
```

Найдем косинусное расстояние от предложения в самой первой строке до всех остальных с помощью функции scipy.spatial.distance.cosine. А также найдем номера двух предложений, ближайших к нему по этому расстоянию.

```python
distances = []

for i in range(1, matrix.shape[0]):
    distances.append(spatial.distance.cosine(matrix[0], matrix[i]))

print(f"Ответ на задание: {distances.index(sorted(distances)[:2][0]) + 1}, {distances.index(sorted(distances)[:2][1]) + 1}")
```

Анализируя данные, можно сделать вывод, что 2 ближайших предложения по вычисленному нами расстоянию совпадают по тематике с первым предложением, а следующие по близости предложения по тематике не совпадают с первым предложением.
