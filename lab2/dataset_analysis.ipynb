С помощью `seaborn.heatmap` построим матрицу корреляции признаков. Для этого необходимо предварительно удалить столбцы с категориальными и бинарными признаками.

Поищим выбросы в данных. Для этого из датасета `c_data` удалим ненужные признаки `product_class_id`, `SKU`. С помощью метода межквантильного промаха найдем нижнию и верхние границы для каждого признака.

````python
r_data = c_data.drop(['product_class_id', 'SKU'], axis=1)
quantiles = {}

for feature in r_data:
    Q1 = r_data[feature].quantile(0.25)
    Q3 = r_data[feature].quantile(0.75)
    IQR = Q3 - Q1
    quantiles.update({feature: [float(Q1 - 1.5 * IQR), float(Q3 + 1.5 * IQR)]})

quantiles
````

Далее создадим такую же по размеру матрицу, как и датасет. Каждая ячейка этой матрицы будет содержать информацию - является ли значение датасета выбросом (0 - нет, 1 - да).

Затем, проходя циклом по каждой строчке матрицы и записывая все в массив `outliers`, будем проверять 2 условия (чтобы проверить, что в строке матрицы все единицы):

1) Состоит ли строка матрицы из одного элемента
2) Является ли первый элемент единицей

Остается создать новый признак у нашего датасета и присвоить ему массив `outliers`. Можно заметить, что выбросов в наших данных нет.

````python
outlier_matrix = np.zeros(r_data.shape)
for feature in r_data:
    feature_min, feature_max = quantiles[feature][0], quantiles[feature][1]
    for cell in r_data[feature]:
        if (cell < feature_min) | (cell > feature_max):
            outlier_matrix = 1

outliers = []
for i in outlier_matrix:
    outliers.append(True if (len(set(i)) == 1) & (i[0] == 1) else False)

data['outlier'] = outliers

data
````
